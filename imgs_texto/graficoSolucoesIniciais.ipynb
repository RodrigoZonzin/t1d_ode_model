```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parmetros da EDO
k = 0.5
m = 2

# Intervalo de tempo
t = np.linspace(0, 10, 100)

# Condies iniciais diferentes
C_vals = np.linspace(-10, 10, 30)

# Criao da grade para t e C
T, C = np.meshgrid(t, C_vals)

# Soluo da EDO
if k != 0:
    Y = (C + m/k) * np.exp(k * T) - m/k
else:
    Y = m * T + C

# Grfico 3D
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(T, C, Y, cmap='viridis', edgecolor='none')
ax.set_xlabel('Tempo t')
ax.set_ylabel('Condio inicial C')
ax.set_zlabel('Soluo y(t)')
ax.set_title('Solues da EDO dy/dt = ky + m para diferentes valores de C')
#ax.view_init(45, 45)

plt.tight_layout()
plt.show()
```
